We want to find the high correlated attributes, these features have a chance to reduce through transformations such as PCA.

```python
#We want to know the continous data's relation.So we ignore the categorical data
data=train_data.iloc[:,116:] 
cols=data.columns 

threshold = 0.5 
corr_list = []

#Search for the highly correlated pairs
for i in range(0,14): #for 'size' features
    for j in range(i+1,14): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) #store correlation and columns index

#Sort to show higher ones first            
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#Print correlations and column names
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))
```

## Convert categorical features to numneric values

```python
#merging both the datasets to make single joined dataset
import pandas
joined = pandas.concat([train_data, test_data],ignore_index = True)
del train_data,test_data                                         #deleting previous one to save memory.

cat_feature = [n for n in joined.columns if n.startswith('cat')]  #list of all the features containing categorical values

#factorizing them
for column in cat_feature:
    joined[column] = pandas.factorize(joined[column].values, sort=True)[0]
        
del cat_feature
```
